## **2D Array**

```python
m=[]
for i in range(7):
  ar=[]
  for j in range(9):
    v=int(input())
    ar.append(v)
  m.append(ar)
print(m)
```

## **Dictionary**

```python
dic={}

for i in range(5):
  k=input()
  li=[]
  for j in range(10):
    if i==0 or i==3:
      v=input()
    elif i==1 or i==4:
      v=int(input())
    elif i==2:
      v=float(input())
    li.append(v)
  dic[k]=li
print(dic)
```

## **Use Import**

```python
import pandas as pd
df=pd.DataFrame(dic)
df
```

## **Encoding**

```python
from sklearn.preprocessing import LabelEncoder

lb=LabelEncoder()
df["For_String"]=lb.fit_transform(df["String"])
df
```
